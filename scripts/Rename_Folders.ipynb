```python
import os
import json

# Load the species ID to name mapping
json_path = "/Users/keira/Desktop/plantnet_300K copy/plantnet300K_species_id_2_name.json"  # Replace with the actual path
with open(json_path, "r") as file:
    species_mapping = json.load(file)  # Example: {"1": "Rose", "2": "Daisy"}

# Define the base dataset directory
dataset_dir = "/Users/keira/Desktop/plantnet_300K copy/images/"  # Replace with the actual path

# Subdirectories: train, test, val
sub_dirs = ["test", "train", "val"]

# Loop through each subdirectory
for sub_dir in sub_dirs:
    folder_path = os.path.join(dataset_dir, sub_dir)
    
    if not os.path.exists(folder_path):
        print(f"Skipping {sub_dir}, folder not found.")
        continue

    # Rename each ID folder to the species name
    for folder_id in os.listdir(folder_path):
        old_folder_path = os.path.join(folder_path, folder_id)
        
        if os.path.isdir(old_folder_path) and folder_id in species_mapping:
            new_folder_name = species_mapping[folder_id]
            new_folder_path = os.path.join(folder_path, new_folder_name)

            # Rename folder
            os.rename(old_folder_path, new_folder_path)
            print(f"Renamed: {folder_id} -> {new_folder_name}")
```

```python
import os

# Define the base dataset directory
dataset_dir = "/Users/keira/Desktop/plantnet_300K copy/images/"  # Replace with your actual path

# Subdirectories: train, test, val
sub_dirs = ["test", "train", "val"]

# Dictionary to store plant image counts per folder
plant_counts = {}

# Set to store unique plant names
unique_plants = set()

# Loop through each subdirectory
for sub_dir in sub_dirs:
    folder_path = os.path.join(dataset_dir, sub_dir)
    
    if not os.path.exists(folder_path):
        print(f"Skipping {sub_dir}, folder not found.")
        continue

    # Loop through each plant folder
    for plant_folder in os.listdir(folder_path):
        plant_folder_path = os.path.join(folder_path, plant_folder)

        # Check if it's a directory
        if os.path.isdir(plant_folder_path):
            unique_plants.add(plant_folder)  # Add to unique set

            # Count the number of images in the folder
            num_images = len([img for img in os.listdir(plant_folder_path) if os.path.isfile(os.path.join(plant_folder_path, img))])

            # Store in dictionary
            if plant_folder not in plant_counts:
                plant_counts[plant_folder] = []
            plant_counts[plant_folder].append((sub_dir, num_images))

# Display results
print(f"\nTotal Unique Plant Species: {len(unique_plants)}\n")

for plant, counts in plant_counts.items():
    print(f"Plant: {plant}")
    for location, num_images in counts:
        print(f"  - {location.capitalize()} Set: {num_images} images")
    print("-" * 30)
```

```python
import os
import csv

# Define the base dataset directory
dataset_dir = "/Users/keira/Desktop/plantnet_300K copy/images/"  # Replace with your actual path

# Subdirectories: train, test, val
sub_dirs = ["test", "train", "val"]

# Dictionary to store plant image counts per folder
plant_counts = {}

# Set to store unique plant names
unique_plants = set()

# Loop through each subdirectory
for sub_dir in sub_dirs:
    folder_path = os.path.join(dataset_dir, sub_dir)
    
    if not os.path.exists(folder_path):
        print(f"Skipping {sub_dir}, folder not found.")
        continue

    # Loop through each plant folder
    for plant_folder in os.listdir(folder_path):
        plant_folder_path = os.path.join(folder_path, plant_folder)

        # Check if it's a directory
        if os.path.isdir(plant_folder_path):
            unique_plants.add(plant_folder)  # Add to unique set

            # Count the number of images in the folder
            num_images = len([img for img in os.listdir(plant_folder_path) if os.path.isfile(os.path.join(plant_folder_path, img))])

            # Store in dictionary
            if plant_folder not in plant_counts:
                plant_counts[plant_folder] = []
            plant_counts[plant_folder].append((sub_dir, num_images))

# Display results
print(f"\nTotal Unique Plant Species: {len(unique_plants)}\n")

# Save results to a CSV file
csv_filename = "plant_image_counts.csv"
with open(csv_filename, mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Plant Name", "Dataset", "Number of Images"])  # Header row

    for plant, counts in plant_counts.items():
        for location, num_images in counts:
            writer.writerow([plant, location, num_images])  # Write each row to CSV

print(f"\nResults saved to {csv_filename}")
```
